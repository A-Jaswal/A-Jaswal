```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("student.csv")
```

```python
df.head()
```

```python
df.rename(columns={'mark': 'score'}, inplace=True)
df
```

```python
import numpy as np

conditions = [
    df['score'] >= 85,
    df['score'].between(70, 84),
    df['score'].between(60, 69),
    df['score'] < 60
]

grades = ['A', 'B', 'C', 'D']

df['grade'] = np.select(conditions, grades, default='D')
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
plt.hist(df['score'], bins=10, color='skyblue', edgecolor='black')
plt.title("Score Distribution")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(x='gender', y='score', data=df)
plt.title("Scores by Gender")
plt.show()
```

```python
avg_grade = df.groupby('grade', as_index=False)['score'].mean()
sns.barplot(x='grade', y='score', data=avg_grade)
plt.title("Average Score by Grade")
plt.show()
```
